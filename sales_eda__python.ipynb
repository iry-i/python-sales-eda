```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'order_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'order_date': [
        '2024-01-05', '2024-01-10', '2024-01-15', '2024-02-01',
        '2024-02-10', '2024-02-15', '2024-03-01', '2024-03-05'
    ],
    'country': ['USA', 'Germany', 'USA', 'UK', 'China', 'Germany', 'USA', 'China' ],
    'category': ['Electronics', 'Electronics', 'Furniture', 'Furniture',
                 'Electronics', 'Furniture', 'Electronics', 'Electronics'],
    'product': ['Laptop', 'Phone', 'Desk', 'Chair', 'Laptop', 'Desk', 'Phone', 'Laptop'],
    'quantity': [1, 2, 1, 4, 1, 2, 3, 1],
    'price': [800, 500, 300, 120, 800, 300, 500, 800]
}
df = pd.DataFrame(data)
df

df['order_date'] = pd.to_datetime(df['order_date'])
df['revenue'] = df['quantity'] * df['price']
df

df['revenue'].sum()
df.groupby('country')['revenue'].sum().sort_values(ascending=False)
df.groupby('category')['revenue'].sum()

revenue_by_country = df.groupby('country')['revenue'].sum()

revenue_by_country.plot(kind='bar')
plt.title('Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.show()

revenue_by_category = df.groupby('category')['revenue'].sum()

revenue_by_category.plot(kind='bar')
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.show()

df['month'] = df['order_date'].dt.to_period('M')

revenue_by_month = df.groupby('month')['revenue'].sum()

revenue_by_month.plot()
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()

"""
Key Insights:

1. USA is the top revenue-generating country.
2. Electronics is the highest-performing product category.
3. Revenue shows a positive trend over time.
"""
```
